```python
import pandas as pd
import os

# Definition of paths relative to the notebook
base_dir = ".."  # go to project root
neg_path = os.path.join(base_dir, "data", "raw", "concepts_negative_subject_final.csv")
pos_path = os.path.join(base_dir, "data", "raw", "concepts_positive_subject_final.csv")
out_path = os.path.join(base_dir, "data", "processed", "merged_questionnaires.csv")

df_neg = pd.read_csv(neg_path)
df_pos = pd.read_csv(pos_path)

# Add diagnosis column
df_pos["Diagnosis"] = 3   # positive
df_neg["Diagnosis"] = 0   # negative

# whitespace and capitalize column names (subject and Subject are used)
df_neg.columns = df_neg.columns.str.strip().str.capitalize()
df_pos.columns = df_pos.columns.str.strip().str.capitalize()

# Merge
df = pd.concat([df_neg, df_pos], ignore_index=True)

# Binarize all questionnaire scores 0 -> 0 ; 1-2-3 -> 1
def binarize(x):
    return 0 if x == 0 else 1

df_binarized = df.applymap(lambda x: binarize(x) if str(x).isdigit() else x)

# Save
os.makedirs(os.path.dirname(out_path), exist_ok=True)
df_binarized.to_csv(out_path, index=False)

print(f"Saved merged + binarized dataset to {out_path}")
```
